```
import matplotlib.pyplot as plt
plt.plot([50,100,1000], [1, 1, 1] , 's-', linewidth=1.5, markersize=3,label="1 hilo")
plt.plot([50,100,1000], [1.971377951, 1.988905328, 1.981585579] , 's-', linewidth=1.5, markersize=3,label="2 hilos")
plt.plot([50,100,1000], [2.064087169, 2.068925439, 2.066107694] , 's-', linewidth=1.5, markersize=3,label="4 hilos")
plt.plot([50,100,1000], [1.983623021, 2.013069926, 1.981085614] , 's-', linewidth=1.5, markersize=3,label="8 hilos")
plt.ylabel('SpeedUp')
plt.xlabel('Iteraciones')
plt.title('Iteraciones vs SpeedUp')
plt.legend()
plt.show()
```

```
plt.plot([1,2,4,8], [1, 1.971377951, 2.064087169, 1.983623021] , 's-', linewidth=1.5, markersize=3,label="50 iteraciones")
plt.plot([1,2,4,8], [1, 1.988905328, 2.068925439, 2.013069926] , 's-', linewidth=1.5, markersize=3,label="100 iteraciones")
plt.plot([1,2,4,8], [1, 1.981585579, 2.066107694, 1.981085614] , 's-', linewidth=1.5, markersize=3,label="1000 iteraciones")
plt.ylabel('SpeedUp')
plt.xlabel('Número de Hilos')
plt.title('Número de Hilos vs SpeedUp')
plt.legend()
plt.show()
```
